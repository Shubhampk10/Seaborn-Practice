```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Join Plot

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='reg')
#kind='reg' or 'hex'
```

## Pair Plot

```python
sns.pairplot(df,hue='sex')
```

## Dist Plot

```python
sns.distplot(df['tip'])
#(df['tip'],5,False) - press shift+tab
```

```python
sns.distplot(df['tip'],kde=False)
```

## For Categorical Plots

## Count Plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

## Bar Plot

```python
sns.barplot(x='total_bill',y='sex',data=df,hue='smoker')
```

```python
sns.barplot(x='total_bill',y='sex',data=df,palette='rainbow')
#palette='Blues_d'/'rocket'/'vlag'
```

## Violin Plot

```python
sns.violinplot(x='total_bill', y='smoker',data=df,palette='rainbow')
```
